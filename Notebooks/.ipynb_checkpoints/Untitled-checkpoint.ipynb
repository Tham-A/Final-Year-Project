```python
import pandas as pd
import numpy as np
import itertools

df = pd.read_csv('LeagueofLegends.csv', converters={'golddiff': pd.eval})


lck2017summerskt = df[(df['League'] == 'LCK') & (df['Year'] == 2017) & (df['Season'] == 'Summer') & ((df['blueTeamTag'] == 'SKT') | (df['redTeamTag'] == 'SKT'))]

arr = np.array(lck2017summerskt['golddiff'], object)
print(arr)
print("***********************************************************************************")
skt2017S = np.array(list(itertools.zip_longest(*arr,fillvalue=np.nan)))
print(skt2017S)
print("***********************************************************************************")
skt2017S = np.nanmean(skt2017S, axis=1)
print(skt2017S)
print("***********************************************************************************")
# test = lck2017summerskt['golddiff']
# print(test)
# print("Data type is %s", type(test))
# test1 = lck2017summerskt['golddiff']
# print(test1)
# print("Data type is %s", type(test1))
```
